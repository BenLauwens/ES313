## GR
#### Random values in a heatmap:

```julia
A = rand(0:10,5,10)
using Plots 
Plots.gr()
```

```julia
Plots.heatmap(A,aspectratio=1,clims=(-5,15),colorbar_title="color legend")
Plots.xticks!(1:2:10,["h tile $i" for i in 1:2:10])
Plots.yticks!(1:5,["v tile $i" for i in 1:5])
plot!(size=(800,400))
title!("heatmap example")
```

```julia
Plots.heatmap(A,aspectratio=1,clims=(-5,15),colorbar_title="color legend",color=:plasma)
Plots.xticks!(1:2:10,["h tile $i" for i in 1:2:10])
Plots.yticks!(1:5,["v tile $i" for i in 1:5])
plot!(size=(800,400))
title!("heatmap example (other colormap)")
```

```julia
Plots.heatmap(A,aspectratio=1,clims=(-5,15),colorbar_title="color legend",color=:plasma_r)
Plots.xticks!(1:2:10,["h tile $i" for i in 1:2:10])
Plots.yticks!(1:5,["v tile $i" for i in 1:5])
plot!(size=(800,400))
title!("heatmap example (reversed the colormap)")
```

```julia
# overview ov available heatmaps:
print(clibraries())
showlibrary(:Plots)
```

Another view using a contour plot instead (using the fill option, otherwise you get more of a topographical map layout.

```julia
Plots.contour(A,fill=true,aspectratio=1,clims=(-5,15))
```
