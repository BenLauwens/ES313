## More problems
### 1
\begin{equation}
\min 3x_1+8x_2+4x_3\\
\textrm{subject to} \begin{cases}
x_1+x_2 \ge 8\\
2x_1-3x_2 \le 0\\
x_2 \ge 9 \\
x_1,x_2 \ge 0
\end{cases}
\end{equation}
### 2
\begin{equation}
\max 3x_1+2x_2-x_3+x_4\\
\textrm{subject to} \begin{cases}
x_1 + 2x_2 + x_3 - x_4 \le 5\\
-2x_1 - 4 x_2 + x_3 + x_4 \le -1\\
x_1\ge 0,x_2 \le 0
\end{cases}
\end{equation}

### 3
\begin{equation}
\min x_1 - x_2 + x_3\\
\textrm{subject to} \begin{cases}
x_1 + 2x_2 - x_3 \le 3\\
-x_1 + x_2 + x_3 \ge 2\\
x_1 - x_2 = 10\\
x_1 \ge 0, x_2 \le 0
\end{cases}
\end{equation}
